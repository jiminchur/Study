```python
# AMEX
# ATEK.U -> ATEK_u.k
# PCG PR A -> PCG_pa
amex_lst = []
for i in tqdm(AMEX):
    if " " in i:
        if i[-2] == " ":
            a,b,c = i.split(" ")
            b_ = b.lower()[0]
            c_ = c.lower()[0]
            code = a + "_" + b_ + c_
            amex_lst.append(code)
        else:
            a,b = i.split(" ")
            b_ = b.lower()[0]
            code = a + "_" + b_
            amex_lst.append(code)
    elif "." in i:
        a,b = i.split(".")
        b_ = b.lower()[0]
        code = a + "_" + b_ + ".k"
        amex_lst.append(code)
    else:
        amex_lst.append(i)
```

```python
# NYSE
# BRK.B -> BRKb
# NI PR B -> NI_pb
nyse_lst = []
for i in tqdm(NYSE):
    if " " in i:
        if i.count(" ")>=2:
            a,b,c = i.split(" ")
            b_ = b.lower()[0]
            c_ = c.lower()[0]
            code = a + "_" + b_ + c_
            nyse_lst.append(code)
        else:
            a,b = i.split(" ")
            b_ = b.lower()[0]
            code = a + "_" + b_
            nyse_lst.append(code)
    elif "." in i:
        a,b = i.split(".")
        b_ = b.lower()[0]
        code = a + b_
        nyse_lst.append(code)
    else:
        nyse_lst.append(i)
```

```python
for i in tqdm(nyse_lst):
    
    url = f"https://api.stock.naver.com/chart/foreign/item/{i}/day?startDateTime=197001010000&endDateTime=202401151122"
    payload = {
    "startDateTime": "197001010000",
    "endDateTime": "202401151115"
    }
    r = requests.get(url, data=payload)
    r_ = r.json()
    df_hk = pd.DataFrame(r_)
    df_hk.to_csv(f"./NYSE_stock/{i}.csv",index=False,encoding="utf-8-sig")
```
